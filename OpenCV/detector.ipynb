```python
import cv2
import tensorflow as tf
from tensorflow.keras.models import Model,Sequential
import numpy as np
from tensorflow.keras.models import  load_model
import os
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
model = load_model("../model_2_MobileNet.h5")
```

```python
def imageDetectior(img, labelName):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img)

    img = cv2.rectangle(img, (40, 40), (400, 300), (255, 30,90), 2)  # (0,255,0),5 color and size
    img = cv2.putText(img, labelName, (35, 35), cv2.FONT_HERSHEY_SIMPLEX, 0.9 ,(255, 0,0), 2)  # 1 is font size
    plt.imshow(img)  
    plt.savefig('Detected {}.jpg'.format(labelName))
    plt.show() 
```

```python

from tensorflow.keras.preprocessing import image
classes = ['daisy', 'dandelion', 'roses', 'sunflowers', 'tulips']
target_path = "sunflowers.jpg"
##############input############## 
img = image.load_img(target_path, target_size=(224,224))
img_fun = cv2.imread(target_path)
img_fun = cv2.resize(img_fun,(470,400))
img_array = image.img_to_array(img)/255
img_batch = np.expand_dims(img_array, axis=0)
prediction = model.predict(img_batch)
max_ind = np.argmax(prediction)
labelName = classes[max_ind]
imageDetectior(img_fun,labelName)

# print(realIndex)
# print(predictionIndex)
```
